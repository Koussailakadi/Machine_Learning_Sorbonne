```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
L=[[1 for i in range(18)],[130,123,98,178,134,100,78,100,95,78,110,130,73,87,152,128,180,124]]


x=np.asarray(L).T
w=np.zeros((2,1))

y=[4805,3783,2689,5683,3750,2684,1478,1685,1980,1075,2423,4253,1754,1873,3487,2954,4762,3076]

y=np.asarray(y).reshape(len(y),1)

XTX_inv=np.linalg.inv(np.matmul(x.T,x))
Xcroix=np.matmul(XTX_inv,x.T)

w=np.dot(Xcroix,y)
print("w=",w)


print("XTX_inv=",x.shape)
print("Xcoix=",Xcroix.shape)
```

```python
print("x=",x.shape)
print("w=",w.shape)
print("y=",y.shape)

x_test=[1,105]
x_test=np.asarray(x_test).reshape(2,1).T


y_predict=np.dot(x_test,w)


plt.figure
plt.scatter(x_test[0][1],y_predict[0][0])
plt.scatter(x[:,1],y)

print("w= ",w)
```

```python
y_predict=np.zeros((18,1))

for i,test in enumerate(x):
    test=test.reshape(1,2)
    y_predict[i]=np.dot(test,w)


#plt.figure()
#plt.scatter(x[:,1],y_predict)
#plt.scatter(x[:,1],y)
#plt.grid()

plt.figure()
plt.plot(y_predict,label="Å·")
plt.plot(y,label="y")
plt.legend()
plt.grid()

c=(y-y_predict)**2   
#calcul de l'erreur total : 
erreur=np.sum(c)
print("erreur SSE=",erreur)
```
